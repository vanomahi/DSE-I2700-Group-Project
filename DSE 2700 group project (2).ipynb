## Observation (2)
There are a lot of unique values in the Victims Age Group attribute. Luckily, the majority of the instances belong in one of the following age groups: <18, 45-64, 18-24, 25-44, UNKNOWN, 65+. 

The instances that don't fall in one of the categories above follow a distinct pattern which are the following 
- (1) have '-' as it's first element 
    - followed by a number which is at most 3 digit long
- (2) doesn't '-' 
    - followed by a number which is at least 3 digits long 
    
We wrote a simple function called clean that took advantage of this pattern and placed the odd instances to one of the six age groups. 

Side note: We should absorb the null values in the unknown category for the age group. Hence, impute the null values to 'UNKNOWN' for this attribute.  

### Decision
- Fill in the null values with 'UNKNOWN'
- Use clean function to categorize age groups
- 6 age groups 
    - <18
    - 45-64
    - 18-24
    - 25-44
    - 65+
    - UNKNOWN

```python
# Note: Value_counts ignore null values
# Most frequent isntaces
NYC_crime['VIC_AGE_GROUP'].value_counts()[:10]
```

```python
# Impute null values to unkown 
NYC_crime.loc[:,'VIC_AGE_GROUP'].fillna('UNKNOWN', inplace=True)
```

```python

def clean(x):
    """
    x: string
    return: one of the six age group as a string 
    """
    check = ['<18', '45-64', '18-24', '25-44','UNKNOWN', '65+']
    if x in check:
        return x
    
    elif x[0] == '-':
        x = x[1:]
        len_of_x = len(x)
        
        if len_of_x >= 3:
            return 'UNKNOWN'
        else:
            int_x = int(x)
            if int_x >= 65:
                return '65+'
            
            elif int_x >= 45 and int_x <= 64:
                return '45-64'
            
            elif int_x >= 25 and int_x <= 44:
                return '25-44'
            
            elif int_x >= 18 and int_x <= 24:
                return '18-24'
            
            else:
                return '<18'
            
    elif len(x) >= 3:
        return 'UNKNOWN'   
```

```python
# check function
print(clean('<18')) # '<18'
print(clean('-978')) # 'UNK'
print(clean('-70')) # '65+'
print(clean('999')) # 'UNK'
print(clean('-4')) # '<18'
```

```python
# Apply function
NYC_crime['VIC_AGE_GROUP'] = NYC_crime.loc[:, 'VIC_AGE_GROUP'].apply(clean)
# Check unique values 
NYC_crime['VIC_AGE_GROUP'].unique()
```

## Attempt to FIx Date
- The dates are not messed up at all

CMPLNT_FR_DT - the date at which the crime occurred or started (this date can be before 2005)

RPT_DT - the date at which the crime was reported (this date has to be after 2005)

```python
"02/11/2015".split('/')
```

```python
trash = []
index = -1
for i in NYC_crime['CMPLNT_FR_DT']:
    index += 1
    if i is not np.nan:
        date = i.split('/')
        year = int(date[-1])
        if year < 2005:
            print(index, year)
            trash.append((index, year))
```

```python
bd = ['1015','1016','1017','1018', '1019', '1025', '1026','1027', '1028','1029']

def clean_dates(x):
    if x is not np.nan:
        date = x.split('/')
        if date[-1] not in bd:
            return x
        else:

            if date[-1] in ['1015','1025']:
                date[-1] = '2015'
                return '/'.join(date)

            elif date[-1] in ['1016','1026']:
                date[-1] = '2016'
                return '/'.join(date)

            elif date[-1]in ['1017','1027']:
                date[-1] = '2017'
                return '/'.join(date)

            elif date[-1] in ['1018', '1028']:
                date[-1] = '2018'
                return '/'.join(date)
            else:
                date[-1] = '2019'
                return '/'.join(date)
    
```

```python
NYC_crime.loc[:, ['PATROL_BORO', 'BORO_NM']].head(20)
```

```python
NYC_crime['CMPLNT_FR_DT'] = NYC_crime.loc[:, 'CMPLNT_FR_DT'].apply(clean_dates)
```
